```python
import pandas as pd
import sqlite3
import numpy as np
```

```python
# connect to database
conn = sqlite3.connect('./data/data_sepatu.db')
conn.row_factory = lambda cursor, row: row[0]
cur = conn.cursor()
```

```python
df_shoe =  pd.read_csv("./data/footware.csv")
df_shoe.head(2)
```

```python
df_shoe.info()
```

```python
df_shoe.values.tolist()[0]
```

```python
new_list = []
# 7 for name and brand
# 8 for price
# 9 list size
for data in df_shoe.values.tolist():
    for val in data[9].split(", "):
        name = " ".join(data[7].split()[1:])
        new_list.append([data[0],data[1],data[5],data[7].split()[0],name,data[8],val,data[11]])

new_list
```

```python
new_df = pd.DataFrame(new_list,columns=['product_id','gender','color','brand','name','price','size','image'])
new_df
```

```python
# df_shoe = new_df.groupby('name',as_index=False)['product_id','name','brand','color','gender','image']
# df_shoe.agg(lambda x: set(x))

df_shoe = new_df.drop_duplicates(subset = "name")[['product_id','name','brand','color','gender','price','image']]
df_shoe
```

```python

df_items = new_df[['product_id','size']]

df_items['stock'] = np.random.randint(0, 9, df_items.shape[0])
```
